#STEP 3 :- Data Preprocessing


Data preprocessing is the process of cleaning and transforming raw data into a format suitable for modeling. It improves model performance and ensures reliable predictions

###Encoding Categorical Variables
Machine learning models cannot directly understand categorical (text) data.
Therefore, we need to convert categorical columns into numerical values.

In this project, we are using Label Encoding with Target Column.

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df["Target"] = le.fit_transform(df["Target"])
```

```python
df["Target"].value_counts()
```

```python
#Heatmap - Correlation Heatmap Of Numerical Features
#feature selection
corr_matrix = df.corr(method="pearson")
plt.figure(figsize=(10, 10))
sns.heatmap(corr_matrix, vmin=-1., vmax=1., annot=False, fmt='.2f', cmap="YlGnBu", cbar=True, linewidths=0.5)
plt.title("Pearson correlation")
plt.show()
```

The correlation heatmap highlights strong relationships between academic performance indicators such as grades and dropout likelihood, while demographic features show weaker correlations. This suggests that academic results play a more significant role in predicting student outcomes.

As we are predicting whether a student will dropout or not so, the number of "Enrolled" student is irrelevant. We only need to know whether a student graduated or dropedout. So, we are dropping the "Enrolled" values and going forward with "Graduate" & "Dropout" values.

```python
df.drop(df[df["Target"]==1].index,inplace=True)
df
```

```python
#Creating a new column which name as Dropout
df["Dropout"] = df["Target"].apply(lambda x: 1 if x==0 else 0)
df
```

###Splitting Features & Lables

```python
X = df.drop(columns=["Target","Dropout"],axis=1)
X.head()
```

#STEP 4 :- Feature Scaling

The dataset contains numerical features with different ranges.
Without scaling, features with larger values may dominate smaller ones and affect model performance.

To overcome this, we will apply Standard scaling with Mean = 0, Std = 1 so that all numerical features are brought to a similar scale.
This improves the training process and ensures fair contribution of each feature.

```python
from sklearn.preprocessing import StandardScaler
X = StandardScaler().fit_transform(X)
```

```python
X = pd.DataFrame(X, columns=df.drop(columns=["Target","Dropout"]).columns)
```

```python
Y = df["Dropout"].values
Y
```

####Selecting Best 5 Important Features By using SelectKBest Method.

```python
from sklearn.feature_selection import SelectKBest, f_classif

selector = SelectKBest(score_func=f_classif, k=5)

X_new = selector.fit_transform(X, Y)

selected_features = X.columns[selector.get_support()]
print("Selected Important Features:")
print(selected_features)

# Agar baad mein selected features ko use karna ho toh
X_selected = X[selected_features]
```

#STEP 5 :- Train-Test Split

```python
from sklearn.model_selection import train_test_split
X_selected_train,X_selected_test,Y_train,Y_test = train_test_split(
    X_selected,Y,
    test_size=0.2,
    random_state=42
)
```

```python
print("Shape of training data : ",X_selected_train.shape)
print("Shape of testing data : ",X_selected_test.shape)
```

#STEP 6 :- Evaluating Multiple Models.

```python
#LogisticRegression Classification Model

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report

LR = LogisticRegression()
LR.fit(X_selected_train,Y_train)
```

```python
prediction_LR = LR.predict(X_selected_test)
print("Accuracy : ",accuracy_score(Y_test,prediction_LR,))
print("Confusion Matrix : \n",confusion_matrix(Y_test,prediction_LR))
print("Classification Report : \n",classification_report(Y_test,prediction_LR))
```

```python
#RandomForestClassifier Classification model

from sklearn.ensemble import RandomForestClassifier

RF = RandomForestClassifier()
RF.fit(X_selected_train,Y_train)
```

```python
prediction_RF = RF.predict(X_selected_test)
print("Accuracy : ",accuracy_score(Y_test,prediction_RF,))
print("Confusion Matrix : \n",confusion_matrix(Y_test,prediction_RF,))
print("Classification Report : \n",classification_report(Y_test,prediction_RF,))
```

```python
#DecisionTreeClassifier Classification model

from sklearn.tree import DecisionTreeClassifier
DT = DecisionTreeClassifier()
DT.fit(X_selected_train,Y_train)
```

```python
prediction_DT = DT.predict(X_selected_test)
print("Accuracy : ",accuracy_score(Y_test,prediction_DT))
print("Confusion Matrix : \n",confusion_matrix(Y_test,prediction_DT))
print("Classification Report : \n",classification_report(Y_test,prediction_DT))
```

```python
#XGBClassifier Classification model

from xgboost import XGBClassifier
XGB = XGBClassifier()
XGB.fit(X_selected_train,Y_train)
```

```python
prediction_XGB = XGB.predict(X_selected_test)
print("Accuracy : ",accuracy_score(prediction_XGB,Y_test))
print("Confusion Matrix : \n",confusion_matrix(prediction_XGB,Y_test))
print("Classification Report : \n",classification_report(prediction_XGB,Y_test))
```

```python
from sklearn.svm import SVC

svm_model = SVC(kernel='rbf', random_state=42)
svm_model.fit(X_selected_train, Y_train)

# Predictions karo
prediction_svm = svm_model.predict(X_selected_test)
```

```python
print("Accuracy : ",accuracy_score(prediction_svm,Y_test))
print("Confusion Matrix : \n",confusion_matrix(prediction_svm,Y_test))
print("Classification Report : \n",classification_report(prediction_svm,Y_test))
```

#Resources for Students at Dropout Risk

```python
# Step 1: Helpful resources.
resources = {
    'academic': ['https://swayam.gov.in/', 'https://nptel.ac.in/'],
    'financial': ['https://scholarships.gov.in/', 'https://buddy4study.com/']
}
```

The Above Helpful resource dictionary  are useful websites to help students with their studies and money problems. The academic links provide free courses to learn more and improve skills. The financial links help students find scholarships and financial support to pay for their education.

```python
# Step 2: Personalized Recommendation Function
def recommend_resources(features, prediction):
    recs = []
    if prediction == 1:
        if features.loc[0, 'Tuition fees up to date'] == 0:
            recs.append(f"Financial Aid: {resources['financial'][0]}")

        if features.loc[0, 'Curricular units 1st sem (grade)'] < 50 or \
           features.loc[0, 'Curricular units 2nd sem (grade)'] < 50:
            recs.append(f"Academic Support: {resources['academic'][0]}")

    return recs
```

This function checks if a student is predicted to be at dropout risk and then suggests personalized resources. It recommends financial aid if tuition fees are unpaid and academic support if semester grades are low. This helps provide targeted help to students who need it most.

```python
#Trying recommendation resources in manual examples

feature_columns = [
    'Tuition fees up to date',
    'Curricular units 1st sem (approved)',
    'Curricular units 1st sem (grade)',
    'Curricular units 2nd sem (approved)',
    'Curricular units 2nd sem (grade)'
]

#Example
student_data = pd.DataFrame([[
    0,   # Tuition fees up to date (0 = No)
    1,   # Curricular units 1st sem (approved)
    55,  # Curricular units 1st sem (grade)
    1,   # Curricular units 2nd sem (approved)
    60   # Curricular units 2nd sem (grade)
]], columns=feature_columns)

pred = RF.predict(student_data)[0]

recommend_resources(student_data, pred)
```

```python
# Step 2: Example student data (DataFrame format)
student2_data = pd.DataFrame([[
    0,   # Tuition fees up to date (0 = No)
    0,   # Curricular units 1st sem (approved)
    40,  # Curricular units 1st sem (grade)
    0,   # Curricular units 2nd sem (approved)
    45   # Curricular units 2nd sem (grade)
]], columns=feature_columns)

pred2 = RF.predict(student2_data)[0]

recommend_resources(student2_data, pred2)
```

```python
student3_data = pd.DataFrame([[
    1,   # Tuition fees up to date (0 = No)
    1,   # Curricular units 1st sem (approved)
    55,  # Curricular units 1st sem (grade)
    0,   # Curricular units 2nd sem (approved)
    40   # Curricular units 2nd sem (grade)
]], columns=feature_columns)

pred3 = RF.predict(student3_data)[0]

recommend_resources(student3_data, pred3)
```

```python
%%writefile app.py
import streamlit as st
import pickle

st.title("Student Dropout Prediction")

model = pickle.load(open("model.pkl", "rb"))

tuition_paid = st.selectbox("Is tuition fees up to date?", ["Yes", "No"])
first_sem_approved = st.selectbox("Curricular units 1st sem (approved)?", ["Yes", "No"])
first_sem_grade = st.number_input("Curricular units 1st sem (grade)", min_value=0.0, max_value=20.0, step=0.1)
second_sem_approved = st.selectbox("Curricular units 2nd sem (approved)?", ["Yes", "No"])
second_sem_grade = st.number_input("Curricular units 2nd sem (grade)", min_value=0.0, max_value=20.0, step=0.1)

# Convert categorical to numeric
tuition_val = 1 if tuition_paid == "Yes" else 0
first_sem_approved_val = 1 if first_sem_approved == "Yes" else 0
second_sem_approved_val = 1 if second_sem_approved == "Yes" else 0

input_data = pd.DataFrame([[
    tuition_val,
    first_sem_approved_val,
    first_sem_grade,
    second_sem_approved_val,
    second_sem_grade
]], columns=[
    'Tuition fees up to date',
    'Curricular units 1st sem (approved)',
    'Curricular units 1st sem (grade)',
    'Curricular units 2nd sem (approved)',
    'Curricular units 2nd sem (grade)'
])

if st.button("Predict Dropout"):
    prediction = model.predict(input_data)[0]
    if prediction == 0:
        st.success("Student is unlikely to drop out.")
    else:
        st.error("Student is likely to drop out.")

    # Show recommendations
    recs = recommend_resources(input_data, prediction)
    if recs:
        st.write("Personalized Recommended Resources:")
        for rec in recs:
            st.markdown(f"- {rec}")


```
